```python
import os 

paths = ['/content/drive/MyDrive/Project III/img/02. 웹툰(선 기준 우선)/게임애니같은/','/content/drive/MyDrive/Project III/img/02. 웹툰(선 기준 우선)/깔끔하고밝음/', '/content/drive/MyDrive/Project III/img/02. 웹툰(선 기준 우선)/진짜선간결/' ]

cnt = 0 
for path in paths : 
  li = os.listdir(path)

  for filename in li : 
    if filename[-3:] != 'jpg' : 
      pass  
    filepath = path + filename 
    image = cv2.imread(filepath)
    #Speed up processing with grayscale
    gray_image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    faces = classifier.detectMultiScale(gray_image)


    for i, (x,y,w,h) in enumerate(faces) : 
      cnt +=1 
      face_image = image[y:y+h, x:x+w] 
      cv2_imshow(face_image)
      cv2.imwrite(f'/content/drive/MyDrive/Project III/img/02. 웹툰(선 기준 우선)/얼굴인식/face{cnt}.jpg',face_image)
```
